## Эмбеддинги BERT

Для русского языка хорошая бертианская модель --- RuBERT из статьи [Adaptation of Deep Bidirectional Multilingual Transformers for Russian Language](https://arxiv.org/abs/1905.07213) (Yuri Kuratov, Mikhail Arkhipov). Она обучалась на российской Википедии и новостях (для сравнения, Multilingual BERT из оригинальной статьи обучалась на Википедии более чем 100 языков).

Работать с моделями будем с помощью [библиотеки](https://huggingface.co/transformers) Huggingface 🤗 Transformers.

```python
from transformers import BertConfig, BertModel, BertTokenizer
import torch


def get_bert_model(name='DeepPavlov/rubert-base-cased'):
    """
    Инициализирует BERT из Huggingface.
    Можно брать любую модель из https://huggingface.co/models.
    Мы возьмём RuBERT
    """
    config = BertConfig.from_pretrained(name, output_hidden_states=True)
    tokenizer = BertTokenizer.from_pretrained(name)
    model = BertModel.from_pretrained(name, config=config)
    return tokenizer, model


bert_tokenizer, bert_model = get_bert_model()
```

В BERT используется токенизатор WordPiece, то есть там иногда встречаются подслова, а не только слова:

```python
list(bert_tokenizer.vocab.keys())[44000:44020]
```

```python
def get_bert_embeddings(sentence, tokenizer=bert_tokenizer, model=bert_model):
    """
    Достаёт эмбеддинги предложения со всех слоёв модели
    :return: list размера 13 из np.ndarray размера num_tokens*hidden_size (для входа модели и каждого из 12 выходов слоёв)
    """
    encoded_input = tokenizer(sentence, return_tensors='pt')
    all_hidden = model(**encoded_input)['hidden_states']
    all_hidden = [hidden[0][1:-1].detach().numpy() for hidden in all_hidden]
    return all_hidden


print('Число слоёв:', len(get_bert_embeddings('Я люблю банан')))
```

## Эмбеддинги ELMo

Воспользуемся библиотекой DeepPavlov и его [`elmo_embedder`](http://docs.deeppavlov.ai/en/master/apiref/models/embedders.html#deeppavlov.models.embedders.elmo_embedder.ELMoEmbedder).

```python
from deeppavlov.models.embedders.elmo_embedder import ELMoEmbedder


def get_elmo_model(name='http://files.deeppavlov.ai/deeppavlov_data/elmo_ru-wiki_600k_steps.tar.gz'):
    """
    Инициализирует ELMo из DeepPavlov.
    По умолчанию -- модель на русской Википедии.
    Ещё можно:
        * Новости на русском языке http://files.deeppavlov.ai/deeppavlov_data/elmo_ru-news_wmt11-16_1.5M_steps.tar.gz
        * Русский твиттер http://files.deeppavlov.ai/deeppavlov_data/elmo_ru-twitter_2013-01_2018-04_600k_steps.tar.gz
    """
    # в параметрах задаём, какие именно слои хотим видеть
    return ELMoEmbedder(name, elmo_output_names=["word_emb", "lstm_outputs1", "lstm_outputs2"], concat_last_axis=False)


elmo_model = get_elmo_model()
```

```python
import numpy as np


class ElmoTokenizer:
    @staticmethod
    def tokenize(sentence):
        return list(sentence.split())


elmo_tokenizer = ElmoTokenizer()


def get_elmo_embeddings(sentence, model=elmo_model):
    """
    Достаёт эмбеддинги предложения со всех слоёв модели
    :return: list размера 3 из np.ndarray размера num_tokens*hidden_size (для входа модели и каждого из 2 LSTM-выходов)
    """
    sentence = elmo_tokenizer.tokenize(sentence)
    layer_word = np.asarray(elmo_model([sentence])[0]).T
    return [np.vstack(layer_word[layer]) for layer in range(3)]


print(elmo_tokenizer.tokenize('Я очень люблю бананы'))
get_elmo_embeddings('Я очень люблю бананы')[1].shape
```

## Эмбеддинги GPT-2

Возьмём [модель GPT-2](https://github.com/sberbank-ai/ru-gpts) от Сбера. Она [есть](https://huggingface.co/sberbank-ai/rugpt2large) в HuggingFace 🤗 Transformers.

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer


def get_gpt2_model(name="sberbank-ai/rugpt2large"):
    """
    Инициализирует GPT-2 Large от Сбера
    """
    tokenizer = GPT2Tokenizer.from_pretrained(name)
    model = GPT2LMHeadModel.from_pretrained(name, output_hidden_states=True)
    return tokenizer, model


gpt2_tokenizer, gpt2_model = get_gpt2_model()
```

```python
list(gpt2_tokenizer.get_vocab().keys())[10000:]
[gpt2_tokenizer.decode(token) for token in gpt2_tokenizer.encode('Я очень люблю банан')]
```

```python
# Поиграться с GPT-2 можно тут:

def generate(
    text, tok=gpt2_tokenizer, model=gpt2_model,
    do_sample=True, max_length=50, repetition_penalty=5.0,
    top_k=5, top_p=0.95, temperature=1,
    num_beams=None,
    no_repeat_ngram_size=3
):
    input_ids = tok.encode(text, return_tensors="pt")
    out = model.generate(
        input_ids,
        max_length=max_length,
        repetition_penalty=repetition_penalty,
        do_sample=do_sample,
        top_k=top_k, top_p=top_p, temperature=temperature,
        num_beams=num_beams, no_repeat_ngram_size=no_repeat_ngram_size
    )
    return list(map(tok.decode, out))

generate('Александр Сергеевич Пушкин родился в ', num_beams=10)
```

```python
def get_gpt2_embeddings(sentence, tokenizer=gpt2_tokenizer, model=gpt2_model):
    """
    Достаёт эмбеддинги предложения со всех слоёв модели
    :return: list размера 37 из np.ndarray размера num_tokens*hidden_size (для входа модели и каждого из 37 выходов слоёв)
    """
    encoded_input = tokenizer(sentence, return_tensors='pt')
    all_hidden = model(**encoded_input)['hidden_states']
    all_hidden = [hidden[0].detach().numpy() for hidden in all_hidden]
    return all_hidden


print('Число слоёв:', len(get_gpt2_embeddings('Я люблю банан')))
```

## Анизотропия моделей

Коротко идея: в статье утверждается, что чем глубже слой моделей, тем больше косинусная близость между рандомными словами. Иными словами, на глубоких слоях эмбеддинги всего словаря расположены ближе друг к другу и представляют собой узкий конус, причём конус сужается с каждым следующим слоем.

Чтобы проверить это, нам надо для отдельного слоя каждой модели повторить следующий алгоритм авторов:

* Для $1000$ случайных предложений взять по одному случайному слову из этого предложения и положить пару (предложение, слово) в массив $A$.
* Сэмплировать $1000$ пар элементов из массива $A$ и усреднить по ним.

```python
np.random.seed(2036)


def get_model2sentenceword():
    """Возвращает словарь, где будут "массивы А" для каждой модели"""
    model2sentenceword = {}
    for name, model, tokenizer in zip(['bert', 'elmo', 'gpt2'], [bert_model, elmo_model, gpt2_model], [bert_tokenizer, elmo_tokenizer, gpt2_tokenizer]):
        model2sentenceword[name] = []
        for sent_idx, sent in enumerate(sentences_1000):
            word_idx = np.random.choice(len(tokenizer.tokenize(sent)))
            model2sentenceword[name].append((sent_idx, word_idx))
    return model2sentenceword
```

```python
random.seed(1984)


def get_model2wordpair(model2sentenceword, sample_sz):
    """Возвращает словарь, где будут случайные пары слов для каждой модели"""
    model2wordpair = {}
    for name, sentenceword in model2sentenceword.items():
        model2wordpair[name] = []
        while len(model2wordpair[name]) < sample_sz:
            word1 = random.choice(sentenceword)
            word2 = random.choice(sentenceword)
            if word1 != word2 and (word1, word2) not in model2wordpair[name] and (word2, word1) not in model2wordpair[name]:
                model2wordpair[name].append((word1, word2))
    return model2wordpair
```

```python
from tqdm.notebook import tqdm


def get_similarity(word1, word2):
    """
    Вычисляет косинусную близость для каждого слоя
    вход --- результат работы get_embeddings(sent)[word_id]
    """
    return np.array([np.sum(layer1 * layer2) / np.linalg.norm(layer1) / np.linalg.norm(layer2) for layer1, layer2 in zip(word1, word2)])


def calculate_mean_similarity(sample_sz=1000):
    """Финальная функция, возвращающая словарь из средних близостей случайных пар слов"""
    model2sentenceword = get_model2sentenceword()
    model2wordpair = get_model2wordpair(model2sentenceword, sample_sz)

    model2sim = {}

    for name, get_embeddings in zip(['bert', 'elmo', 'gpt2'], [get_bert_embeddings, get_elmo_embeddings, get_gpt2_embeddings]):
        for (sent1, word1), (sent2, word2) in tqdm(model2wordpair[name]):
            sent1 = get_embeddings(sentences_1000[sent1])
            word1 = [layer[word1, :] for layer in sent1]
            sent2 = get_embeddings(sentences_1000[sent2])
            word2 = [layer[word2, :] for layer in sent2]
            if name not in model2sim:
                model2sim[name] = np.zeros(len(word1))
            model2sim[name] += get_similarity(word1, word2)
        model2sim[name] /= sample_sz
    return model2sim
```

```python
model2sim = calculate_mean_similarity()
```

```python
import matplotlib
import matplotlib.pyplot as plt


matplotlib.rc('axes', edgecolor='k')

plt.figure(figsize=(16, 8))
icons = ['ro:', 'bo:', 'go:']

for i, (model, num_layers) in enumerate([('ELMo', 3), ('BERT', 13), ('GPT2', 37)]):
    x = np.array(range(num_layers))
    plt.plot(x, model2sim[model.lower()], icons[i], markersize=6, label=model, linewidth=2.5, alpha=0.65)

plt.grid(True, linewidth=0.25)
plt.legend(loc='upper left', fontsize=14)   
plt.xlabel('Layer Index', size=14)
plt.xticks(x, size=12)
plt.ylim(0, 1.0)
plt.title("Average Cosine Similarity between Randomly Sampled Words", size=20)
# plt.savefig(f'img/mean_cosine_similarity_across_words.png', bbox_inches='tight')
plt.show()
```

Видно, что траектория графика BERT хорошо бьётся с оригинальной статьёй (напоминаю, там это Figure 1), только амплитуда побольше и на последних слоях график чуть смещён вниз относительно английского. Даже видны экстремумы в слоях $6$ и $8$.

Если обрезать GPT2 по $12$-й слой, то траектория GPT2 тоже бьётся (но, напротив, график повыше для русского). Потом близость немного снижается.

Ниже исследование, что за ступенька у GPT2 в конце.

```python
# смотрим модель. Видим в конце слой LayerNorm

gpt2_model
```

Проверяю, какой слой последний в hidden_states:

```python
current_last = get_gpt2_embeddings('Произошёл троллинг')[-1]

# меняю (пред)последний слой (ln_f)
a = gpt2_model.transformer.ln_f
gpt2_model.transformer.ln_f = torch.nn.Identity()

print(current_last - get_gpt2_embeddings('Произошёл троллинг')[-1])  # тут видно, что эмбеддинг поменялся. Поэтому слой ln_f есть в хидден-стейтах

# возвращаю
gpt2_model.transformer.ln_f = a
```

То есть слой `ln_f` входит в `hidden_states`.

В оригинальной статье нет этой ступеньки, потому что там берётся GPT-2 без головы (`GPT2Model`), и поэтому её последний слой (`LayerNorm`) --- это её выход, а не скрытый слой. В коде мы берём `GPT2LMHeadModel` --- модель с головой, поэтому `LayerNorm` считается как скрытый слой и на графике появляется ступенька.

Получается, что несмотря на анизотропность скрытых слоёв GPT-2, сам выход из неё (и последний скрытый слой в головастых моделях) изотропичен.
